```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame

def odes(x,t):
    #constan
    alpha = 0.4
    beta = 0.35
    gamma = 0.3
    r = 0.3
    k = 0.05
    s = 0.2
    f = 0.15 # menggantikan parameter x pada model.
    u = 64
    mu = 0.1
    p = 0.15
    q1 = 0.2
    q2 = 0.2
    q3 = 0.2
    y1 = 0.35
    y2 = 0.05
    

    #vector element
    A = x[0]
    B = x[1]
    C = x[2]
    D = x[3]
    
    # persamaan 
    dAdt = u - (alpha + r + p + mu + k*C)*A + y1*C + q1*D
    dBdt = alpha*A - (beta + s + mu + p + f*D)*B + y2*C + q2*D
    dCdt = r*A + beta*B - (gamma + mu + p + y1 + y2 - k*A)*C + q3*D
    dDdt = s*B + gamma*C - (mu + p + q1 + q2 + q3 - f*B)*D
    
    return (dAdt, dBdt, dCdt, dDdt)

#nilai awal (jumlah populasi awal)
x0=[24,15,11,14]

#grafik waktu
t = np.linspace(0,20,1000)
x = odeint(odes,x0,t)

A = x[:,0]
B = x[:,1]
C = x[:,2]
D = x[:,3]

data={'t':t,'A':A, 'B':B, 'C':C, 'D':D}
print(DataFrame(data))

#plot 
plt.figure(figsize=(13,7))
plt.plot(t,A, label='Mengambil mata kuliah maksimal tahun ke-1', color='green')
plt.plot(t,B, label='Mengambil mata kuliah maksimal tahun ke-2', color='red')
plt.plot(t,C, label='Mengambil mata kuliah maksimal tahun ke-3', color='blue')
plt.plot(t,D, label='Mengambil mata kuliah maksimal tahun ke-4', color='yellow')
plt.xlabel('t', fontsize=18)
plt.ylabel('x', fontsize=16)
plt.legend()
plt.grid()
plt.title("Kasus 1")
plt.savefig('Kasus 1.png',dpi = 500)
plt.show()
```
